```python
## Import libraries
import requests
import bs4 as bs
import pandas as pd
```

First, I scrape Presidential Debates from the Commission of Presidential Debates website: http://www.debates.org/index.php?page=debate-transcripts.

I am not allowed to manually look up the URLs that I need, instead I have to scrape them. The root url to be scraped is the one listed above, namely: http://www.debates.org/index.php?page=debate-transcripts


I write a command to request the data from debates.org and I Store the content in the variable called original

```python
root_url = 'http://www.debates.org/index.php?page=debate-transcripts'
original = requests.get(root_url).content
original
```

I use the original variable to create a BeautifulSoup object.

```python
soup = bs.BeautifulSoup(original, features='html.parser') 
```

```python
## Let's look at what the scraped website looks like
soup.prettify
```

I extract all the a links that contain the words "First", except the debates from 1992 and I store the results in the array called urls. 

```python
urls=[]
for link in soup.find_all('a'):
    if "First" in link.string:
        if "1992" in link.text:
            continue
        if ":" not in link.text:
            continue
        print(link.text)
        urls.append(link)

urls
```

Then, I create an NaN valued dataframe and I assign it to a variable df0. I set only the columns and the index for this dataframe. 

```python
import numpy as np
cols = []

for link in soup.findAll('a'):
    if 'The First' in link.string:
        cols.append(link.string)
               
index_values= ['Debate char length', 'war_count','most_common_w', 'most_common_w_count']
print(cols)

df0 =  pd.DataFrame(np.nan, index=index_values, columns=cols)
df0
```

Now I create the dataframe by going to every link that I captured in the variable urls.

```python
import re
import operator


data = dict()
for idx, url in enumerate(urls):
    real_url = "http://www.debates.org" + url.get("href") 
    page = requests.get(real_url).content
    soup_temp = bs.BeautifulSoup(page, features='lxml') 
    content = soup_temp.find(id= "content-sm").text
    words = dict()
    char_count = 0
    lines = content.split("\n")
    for line in lines:
        char_count += len(line)
        line.split(" ")
        for word in line.split(" "):
            if word.lower() in words:
                words[word.lower()] +=1
            else:
                words[word.lower()] = 1            
    words = {k: v for k, v in sorted(words.items(), key=lambda item: item[1], reverse=True)}
    most_common_w = max(words.items(), key=operator.itemgetter(1))[0]
    most_common_w_count = words[most_common_w]
    x = re.compile(r'[Ww]ars*[\s. ,]')
    matches = x.findall(content)
    war_count = len(matches)
    df0[url.text]["Debate char length"]= char_count
    df0[url.text]["most_common_w_count"]= most_common_w_count
    df0[url.text]["war_count"]= war_count
    df0[url.text]["most_common_w"]= most_common_w

## Print df0
#df0
#df0.loc['most_common_w_count'].sum()
df0
```
